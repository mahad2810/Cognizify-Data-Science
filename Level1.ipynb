```python
import pandas as pd

restaurant_df=pd.read_csv('Dataset.csv')

def invalid_characters(value):
    return any(ord(char) > 127 for char in value)

restaurant_df = restaurant_df[~restaurant_df['City'].apply(invalid_characters)]
restaurant_df = restaurant_df[~restaurant_df['Address'].apply(invalid_characters)]

table_booking = restaurant_df['Has Table booking'].value_counts(normalize=True) * 100
online_delivery= restaurant_df['Has Online delivery'].value_counts(normalize=True) * 100


average_rating_table_booking = restaurant_df.groupby('Has Table booking')['Aggregate rating'].mean()

online_delivery_by_price_range = restaurant_df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True) * 100

common_price = restaurant_df['Price range'].mode()[0]

average_rating_by_price_range = restaurant_df.groupby('Price range')['Aggregate rating'].mean()

highest_avg_rating_color = restaurant_df.groupby('Rating color')['Aggregate rating'].mean().idxmax()

restaurant_df['Restaurant Name Length'] = restaurant_df['Restaurant Name'].str.len()
restaurant_df['Address Length'] = restaurant_df['Address'].str.len()

restaurant_df['Has Table Booking Encoded'] = restaurant_df['Has Table booking'].map({'Yes': 1, 'No': 0})
restaurant_df['Has Online Delivery Encoded'] = restaurant_df['Has Online delivery'].map({'Yes': 1, 'No': 0})



print("Percentage of Restaurants Offering Table Booking:")
print(table_booking)

print("\nPercentage of Restaurants Offering Online Delivery:")
print(online_delivery)

print("\nAverage Ratings (With and Without Table Booking):")
print(average_rating_table_booking)

print("\nOnline Delivery Availability by Price Range:")
print(online_delivery_by_price_range)

print("\nMost Common Price Range:")
print(common_price)

print("\nAverage Rating by Price Range:")
print(average_rating_by_price_range)

print("\nRating Color Representing the Highest Average Rating:")
print(highest_avg_rating_color)


restaurant_df.to_csv('modified_restaurants.csv', index=False)
```
